#### **과적합 이해하기**
>* 과적합(over fitting): 모델이 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것.  
* 층이 너무 많거나 변수가 복잡, 학습셋과 테스트셋이 중복될 때 발생.  
* 그림 넣기(p.160)
* 과적합 방지 방법: more training data, reduce the # of features, regularization(L1, L2, dropout), split dataset(학습셋과 테스트셋의 구분)
* 그림 넣기(p.161, 162,163)
* 은닉층이 늘어날수록 학습셋의 예측률이 올라가지만 어느 순간 테스트셋의 예측률은 떨어짐.  

#### **모델 저장과 재사용**
>* 학습이 끝난 후 모델을 저장하여 새로운 데이터에 사용할 수 있음.
* keras.models의 load_model 사용 


#### **K겹 교차 검증(K-fold cross validation)**
>* 데이터가 충분하지 않을때, 
데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 학습셋으로 사용하는 방법
* 그림 넣기(p.169)
* sklearn.model_selection의 StratifiedKFold() 사용

```
# Library
from keras.models import Sequential, load_model
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from google.colab import files

import pandas as pd
import numpy as np
import tensorflow as tf
import io
```

```
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)
```

```
# 데이터 불러오기
uploaded = files.upload()
df = pd.read_csv(io.StringIO(uploaded['sonar.csv'].decode('utf-8')), header=None)
print(df.shape)
df.head(5)
```

```
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```
# 전체 데이터 이용
# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim = 60, activation='relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(loss = 'mean_squared_error',
              optimizer = 'adam',
              metrics = ['accuracy'])

# 모델 실행
model.fit(X, Y, epochs = 200, batch_size = 5)
model.save('my_model1.h5')

# 결과 출력
print("\n Accuracy: %.4f" %(model.evaluate(X, Y)[1]))
```

```
# 학습셋과 테스트셋의 구분
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = seed)      # train set:test set = 70:30
```

```
# 학습셋으로 모델 생성 후 테스트셋에 모델 적용
# 모델 설정
model2 = Sequential()
model2.add(Dense(24, input_dim = 60, activation = 'relu'))
model2.add(Dense(10, activation = 'relu'))
model2.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model2.compile(loss = 'mean_squared_error',
               optimizer = 'adam',
               metrics = ['accuracy'])

# 모델 실행
model2.fit(X_train, Y_train, epochs = 200, batch_size = 5)
model2.save('my_model2.h5')

# 테스트셋에 모델 적용 결과 출력
print("\n Test Accuracy: %.4f" %(model2.evaluate(X_test, Y_test)[1]))
```

```
del model   # 테스트 위해 메모리 내의 모델 삭제
del model2
model1 = load_model('my_model1.h5')   # 모델 새로 불러옴
model2 = load_model('my_model2.h5')

print("\n Test Accuracy: %.4f" %(model1.evaluate(X_test, Y_test)[1]))
print("\n Test Accuracy: %.4f" %(model2.evaluate(X_test, Y_test)[1]))
```

#### **K겹 교차 검증(K-fold cross validation)**


```
from sklearn.model_selection import StratifiedKFold
# 10개의 파일로 쪼갬
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```
# 빈 accuracy 배열
accuracy = []

# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X, Y):
  model3 = Sequential()
  model3.add(Dense(24, input_dim = 60, activation='relu'))
  model3.add(Dense(10, activation='relu'))
  model3.add(Dense(1, activation='sigmoid'))
  model3.compile(loss = 'mean_squared_error',
                 optimizer = 'adam',
                 metrics = ['accuracy'])
  model3.fit(X[train], Y[train], epochs = 100, batch_size = 5)
  k_accuracy = "%.4f" % (model3.evaluate(X[test], Y[test])[1])
  accuracy.append(k_accuracy)

# 결과 출력
print("\n %.f fold accuracy: " % n_fold, accuracy)
```
