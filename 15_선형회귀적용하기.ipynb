### 15. 선형회귀 적용하기
> 보스턴 집값 예측  
> 수치 예측 문제  
>  마지막에 참/거짓 등 구분할 필요가 없기 때문에 출력층에 활성화 함수를 지정할 필요가 없음

```
# Library
from keras import models, layers
from sklearn.model_selection import train_test_split

from google.colab import files
import numpy as np
import pandas as pd
import tensorflow as tf
import io
```

```
# seed 설정
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)
```

```
# 데이터 로드
uploaded = files.upload()
```

```
housing = pd.read_csv(io.StringIO(uploaded['housing.csv'].decode('utf-8')), delim_whitespace=True, header=None)

# 데이터 살펴보기
print("shape: ",housing.shape)         # (506, 14): 506개의 샘플, 13개 속성 + 1개의 클래스
print("info: ",housing.info())
housing.head()
```

column | 의미
--- | ---
0 | CRIM: 인구 1인당 범죄 발생 수
1 | ZN: 25,000평방 피트 이상의 주거 구역 비중
2 | INDUS: 소매업 외 상업이 차지하는 면적 비율
3 | CHAS: 찰스강 위치 변수(1: 강 주변, 0: 이외)
4 | NOX: 일산화질소 농소
5 | RM: 집의 평균 방 수
6 | AGE: 1940년 이전에 지어진 비율
7 | DIS: 5가지 보스턴 시 고용 시설까지의 거리
8 | RAD: 순환고속도로의 접근 용이성
9 | TAX:$10,000당 부동산 세율 총계 
10 | PTRATIO: 지역별 학생과 교사 비율
11 | B: 지역별 흑인 비율
12 | LSTAT: 급여가 낮은 직업에 종사하는 인구 비율(%)
13 | 가격(단위: 1,000)



```
# 모델링을 위해 속성값과 클래스값으로 분리
dataset = housing.values
X = dataset[:,0:13]
Y = dataset[:,13]

# train, test 셋으로 7:3 비율로 나눠줌
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = seed)
```

```
# 모델 설정
model = models.Sequential()
model.add(layers.Dense(30, input_dim = 13, activation = 'relu'))
model.add(layers.Dense(6, activation = 'relu'))
model.add(layers.Dense(1))  # 마지막에 참/거짓 등 구분할 필요가 없음. 따라서 출력층에 활성화 함수 지정할 필요가 없음

# 모델 컴파일
model.compile(loss = 'mean_squared_error', optimizer = 'adam')    # metric은 학습결과에 영향을 미치는 것이 아님. 학습 과정이 제대로 이루어지는지 확인을 하기 위함.

# 모델 실행
model.fit(X_train, Y_train, epochs = 200, batch_size = 10)
```

```
# 예측값과 실제 값의 비교
Y_prediction = model.predict(X_test).flatten()   # flatten(): 데이터 배열을 1차원으로 바꿔 읽기 쉽게 해 주는 함수
for i in range(10):
  label = Y_test[i]
  prediction = Y_prediction[i]
  print("실제가격: {:.3f}, 예상가격: {:.3f}".format(label, prediction))
```
